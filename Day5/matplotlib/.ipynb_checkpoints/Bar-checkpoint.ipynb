```python
import matplotlib.pyplot as plt
x=["Nepal","India","china"]
y=[1,2,3]
# z=[5,6,8]
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Population of different country",fontsize=30)
# c=['r','g','y']
plt.bar(x,y,width=0.5,color='r',edgecolor='b',linewidth=3,label="Sanjib")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["Nepal","India","china"]
y=[1,2,3]
p=np.arange(len(x))
width=0.2
p1=[width+j for j in p]
z=[5,6,8]
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Population of different country",fontsize=30)
# c=['r','g','y']

plt.bar(p,z,width,color='b',linewidth=3,label="Sanjib")
plt.bar(p1,y,width,color='r',linewidth=3,label="Sanjib")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
