# PIe chart

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]

y=["kathmmandu","kailali","achham","bajura"]
plt.pie(x,labels=y,explode=[0.3,0,0,0],autopct="%0.2f%%",shadow=True,radius=1.5,startangle=90)
plt.title("District")
plt.legend(loc=1)
plt.show
```
