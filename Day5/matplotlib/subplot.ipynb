# subplot abd save file

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4]
y1=np.array([2,3,5,8])
plt.subplot(2,2,1)
plt.plot(x,y1)

y2=y1**2
plt.subplot(2,2,2)
plt.plot(x,y2)

y3=y2**2
plt.subplot(2,2,3)
plt.plot(x,y3)

y4=y3**2
plt.subplot(2,2,4)
plt.plot(x,y4)

plt.savefig("test",dpi=300)
plt.show()
```
