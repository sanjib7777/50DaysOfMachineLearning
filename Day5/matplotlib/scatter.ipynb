Scatter plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,34,4,5,4]
y=[4,5,3,8,3,5,3,3]
color=[30,10,20,30,40,50,60,70]
size=[2,3,4,5,6,7,8,9]
sizes=[x**3 for x in size]
plt.scatter(x,y,c=color,s=sizes,marker="*",cmap="Accent")
plt.colorbar()
plt.title('My scatter',fontsize=20,color='y')
plt.xlabel('Days')
plt.ylabel('No')
plt.show()
```
