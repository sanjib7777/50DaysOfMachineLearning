```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cars.csv")
df.head()
```

```python
df['owner'].unique()
```

```python
df = df[df['brand'] != 'Ashok']
```

```python
df.shape
```

```python
 total = df['brand'].value_counts() 
```

```python
df['brand'].nunique()
```

```python
rep = total[total<=100].index
```

```python
df['brand'] = df['brand'].apply(lambda x:x if total[x]>=100 else "others")
```

```python
df['brand'].value_counts()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(sparse_output=False,drop='first'),['brand','fuel']),
    ('tnf2',OrdinalEncoder(categories=[['First Owner', 'Second Owner', 'Third Owner','Fourth & Above Owner', 'Test Drive Car']]),['owner'])
],remainder='passthrough')
```

```python
X_transformed = transformer.fit_transform(X)
X_transformed = pd.DataFrame(X_transformed) 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_transformed,y,test_size=0.2,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,accuracy_score
```

```python
print(f"R2 Score: {r2_score(y_test, y_pred)}")
```
