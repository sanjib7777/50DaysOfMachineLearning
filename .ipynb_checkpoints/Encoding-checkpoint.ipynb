```python
df=pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
pd.get_dummies(df,columns=['fuel','owner'],dtype=int,drop_first=True)
```

```python
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df[['fuel','owner']],df['selling_price'],test_size=0.2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(sparse_output=False,drop='first')
```

```python
ohe.fit(X_train)
```

```python
X_train_trans = ohe.transform(X_train)
X_test_trans = ohe.transform(X_test)
```

```python
X_train_trans
```

```python
import numpy as np
```

```python
X_train_np = np.hstack((X_train[['fuel','owner']].values,X_train_trans))
X_train_np.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
    ('tnf1',OneHotEncoder(sparse_output=False,drop='first'),['fuel','owner'])
],remainder='passthrough')
```
