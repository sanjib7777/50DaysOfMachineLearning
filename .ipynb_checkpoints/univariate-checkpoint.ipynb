```python
import pandas as pd
```

```python
df=pd.read_csv("ipl.csv")
```

```python
df.head()
```

```python
df=df.drop('umpire3',axis=1)
```

```python
df.head()
```

```python
city=df['city']
```

```python
city=city.value_counts()
```

```python
x=city.index.tolist()
x
```

```python
y=city.values.tolist()
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20))
plt.barh(x,y)
plt.title('Most played city',fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```

```python
POTM_10=df['player_of_match'].value_counts().head(10)
```

```python
y=POTM_10.values.tolist()
```

```python
x=POTM_10.index.tolist()
```

```python
plt.barh(x,y)
plt.title('Most player of the match player')
plt.show()
```
