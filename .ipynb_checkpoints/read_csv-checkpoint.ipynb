```python
import pandas as pd
```

```python
def rename(name):
    if name=="Royal Challengers Bangalore" :
        return "RCB"
    else:
        return name
    
```

```python
pd.read_csv('ipl.csv',nrows=100,dtype={'win_by_wickets':float},parse_dates=['date'],converters={"team1":rename,"team2":rename,"winner":rename})
```
